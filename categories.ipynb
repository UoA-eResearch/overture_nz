```python
import pandas as pd
df = pd.read_csv("overture_NZ.csv")
df
```

```python
df.main_category.value_counts().to_csv("categories.csv")
```

```python
# Add top level category from taxonomy
overture_taxonomy = pd.read_csv("https://raw.githubusercontent.com/OvertureMaps/schema/main/task-force-docs/places/overture_categories.csv", sep=";")
overture_taxonomy["Overture Taxonomy"] = overture_taxonomy["Overture Taxonomy"].apply(lambda s: s.strip("[]").split(","))
overture_taxonomy.set_index("Category code", inplace=True)
overture_taxonomy = overture_taxonomy["Overture Taxonomy"]
df["top_level_category"] = df.main_category.apply(lambda c: overture_taxonomy.get(c, [None])[0])
df
```

```python
df.top_level_category.value_counts()
```

```python
mapping_dict = {}
for line in open("category_mapping.txt", encoding="utf8"):
  if " -> " not in line:
    continue
  overture_categories, resulting_category = line.strip().split(" -> ")
  for c in overture_categories.split(","):
    mapping_dict[c] = resulting_category
mapping_dict
```

```python
df["mapped_category"] = df.main_category.apply(lambda c: mapping_dict.get(c, "other"))
df["mapped_category"].value_counts()
```
